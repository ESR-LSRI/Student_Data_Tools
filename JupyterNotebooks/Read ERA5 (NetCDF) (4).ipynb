```python
#import libraries
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as cc
import cartopy.feature as cp
```

```python
#open data set and view
ds = xr.open_dataset('')
ds
```

```python
#If there are multiple time values, pick a timestamp from time category above and paste here. var is the variable you are graphing
var = ds['var'].sel(time = '')
```

```python
#Assign lat and lon variables
lat = ds.latitude
lon = ds.longitude
```

```python
# Setup the figure
fig = plt.figure(figsize=(20, 10))
ax = plt.subplot(projection=cc.PlateCarree())
plt.title(f"Insert title")

# Plot the data for the variable you want
var = ax.contourf(lon, lat, var, cmap='jet', levels=20)

#Add details
plt.colorbar(var, label='insert title here')
ax.coastlines()

```
